## 💡 변수(Variable)란?

**변수**는 값을 저장하는 **상자**예요! 📦

### 실생활 비유
상자에 물건을 담고, 상자에 라벨을 붙이는 것과 같아요.

```
📦 name (상자)
   ↓
 "김철수" (담긴 내용)
```

### 왜 변수를 쓰나요?

#### 1️⃣ 값을 기억할 수 있어요
```python
# 변수 없이
print("김철수")
print("김철수")
print("김철수")  # 계속 반복...

# 변수로
name = "김철수"
print(name)
print(name)
print(name)  # 한 번만 바꾸면 전부 바뀌어요!
```

#### 2️⃣ 의미를 명확하게 표현해요
```python
print(25)              # 이게 뭐지?
print(age)             # 아, 나이구나!
```

#### 3️⃣ 계산 결과를 저장해요
```python
total = 100 + 200
print(total)           # 300
```

---

````python
# 👉 Practice 1: 변수 만들기
# 변수 생성은 = (등호)를 사용해요
# 왼쪽에 변수명, 오른쪽에 값

name = "김철수"
age = 25
height = 175.5

print("변수 생성 완료!")
print(name)
print(age)
print(height)
````

## 💡 변수 할당(Assignment) 이해하기

### = 기호의 의미
```python
name = "김철수"
```

📌 **주의**: `=`는 "같다"가 아니라 "오른쪽 값을 왼쪽 변수에 저장하라"는 뜻이에요!

### 실행 순서
1. 오른쪽 값을 먼저 확인 → `"김철수"`
2. 왼쪽 변수에 저장 → `name`이라는 상자에 담기

### 변수는 덮어쓸 수 있어요
```python
age = 25      # age에 25 저장
age = 30      # age를 30으로 변경
print(age)    # 30 (최신 값만 남아요)
```

````python
# 👉 Practice 2: 변수 사용하기

# 변수 만들기
name = "김철수"
age = 25
height = 175.5

# 변수로 계산하기
next_year_age = age + 1
print(f"내년 나이: {next_year_age}")

# 변수 값 변경하기
age = 26
print(f"변경된 나이: {age}")

# 변수끼리 연산하기
double_age = age * 2
print(f"나이의 두 배: {double_age}")
````

## 📝 변수명 작성 규칙

### ✅ 지켜야 하는 규칙 (안 지키면 에러!)

1. **문자, 숫자, 언더스코어(_)만 사용**
   - ✅ `name`, `age_1`, `user_name`
   - ❌ `name!`, `age@`, `user-name`

2. **숫자로 시작 불가**
   - ✅ `name1`, `user2`
   - ❌ `1name`, `2user`

3. **예약어 사용 불가**
   - ❌ `if`, `for`, `while`, `class`, `def` 등
   - 이건 Python이 이미 쓰고 있는 단어예요

4. **대소문자 구분**
   - `name`, `Name`, `NAME`은 다른 변수!

### 🎨 권장하는 스타일 (Python 관례)

#### snake_case 사용 (Python 스타일!)
- 모든 글자는 소문자
- 단어 사이는 언더스코어(_)로 연결

```python
✅ user_name = "김철수"
✅ total_price = 10000
✅ is_valid = True

❌ userName (JavaScript 스타일)
❌ TotalPrice (다른 언어 스타일)
```

#### 의미 있는 이름 짓기
```python
✅ student_count = 30
❌ sc = 30  (뭔지 모르겠어요)

✅ total_price = 15000
❌ tp = 15000  (헷갈려요)
```

## 🔍 type() 함수로 타입 확인하기

변수에는 **타입(type)**이 있어요!

### 타입이란?
데이터의 **종류**를 말해요.

```
📦 name = "김철수"  → 문자열(str) 타입
📦 age = 25         → 정수(int) 타입
📦 height = 175.5   → 실수(float) 타입
📦 is_student = True → 불린(bool) 타입
```

### type() 함수
변수가 어떤 타입인지 알려줘요!

```python
type(변수)
```

### 왜 타입이 중요한가요?
- 타입이 다르면 할 수 있는 일이 달라요
- 문자열은 더하면 연결되고, 숫자는 더하면 계산돼요
- 나중에 FastAPI에서 타입이 엄청 중요해요!

---

````python
# 👉 Practice 4: type() 함수로 타입 확인하기

name = "김철수"
age = 25
height = 175.5
is_engineer = True

print(type(name))           # <class 'str'> - 문자열
print(type(age))            # <class 'int'> - 정수
print(type(height))         # <class 'float'> - 실수
print(type(is_engineer))    # <class 'bool'> - 불린
````

## 📚 Python의 주요 타입들

### 1️⃣ str (String, 문자열)
- 글자, 문장을 저장
- 따옴표로 감싸요: `"안녕"`, `'hello'`

```python
name = "김철수"
message = 'Hello!'
```

### 2️⃣ int (Integer, 정수)
- 소수점 없는 숫자
- 나이, 개수, 점수 등

```python
age = 25
count = 100
score = 95
```

### 3️⃣ float (Floating Point, 실수)
- 소수점 있는 숫자
- 키, 몸무게, 온도 등

```python
height = 175.5
weight = 70.3
temperature = 36.5
```

### 4️⃣ bool (Boolean, 불린)
- 참(True) 또는 거짓(False)만 저장
- 첫 글자는 대문자!

```python
is_student = True
is_raining = False
```

---

````python
# 👉 Practice 5: 다양한 타입의 변수 만들기

# 문자열 (str)
company = "네이버"
language = "Python"

# 정수 (int)
employee_count = 3500
year = 2025

# 실수 (float)
average_salary = 5500.5
stock_price = 123.45

# 불린 (bool)
is_hiring = True
is_public = False

# 타입 확인
print(f"{company}의 타입: {type(company)}")
print(f"{employee_count}의 타입: {type(employee_count)}")
print(f"{average_salary}의 타입: {type(average_salary)}")
print(f"{is_hiring}의 타입: {type(is_hiring)}")
````

## 🏷️ 타입 힌트(Type Hint) 미리보기

Python에서는 변수의 타입을 **명시**할 수 있어요!

### 기본 형태
```python
변수명: 타입 = 값
```

### 예시
```python
name: str = "김철수"
age: int = 25
height: float = 175.5
```

### 타입 힌트가 뭐가 좋나요?

#### 1️⃣ 코드가 명확해져요
```python
# 타입 힌트 없이
data = "hello"  # 이게 뭐지?

# 타입 힌트로
user_name: str = "hello"  # 아, 문자열이구나!
```

#### 2️⃣ VS Code가 도와줘요
- 자동 완성이 더 잘 돼요
- 오타를 미리 잡아줘요

#### 3️⃣ FastAPI에서 필수!
- FastAPI는 타입 힌트를 많이 사용해요
- 지금부터 익숙해지면 나중에 편해요!

### ⚠️ 주의사항
타입 힌트는 **힌트일 뿐**이에요. 강제가 아니에요!

```python
age: int = "25"  # 에러가 안 나요! (하지만 권장하지 않아요)
```

---

````python
# 👉 Practice 6: 타입 힌트 사용하기

# 타입 힌트 없이
name = "김철수"
age = 25
height = 175.5

# 타입 힌트로 (명확하죠?)
name: str = "김철수"
age: int = 25
height: float = 175.5
is_student: bool = True

print(name)
print(age)
print(height)
print(is_student)

# VS Code에서 name. 찍어보면 자동 완성이 나와요!
# (문자열 메서드들이 추천돼요)
````
