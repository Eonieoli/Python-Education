## 📝 주석(Comment) 사용법

**주석**은 코드에 메모를 남기는 거예요. Python이 실행하지 않는 부분이죠.

### 왜 주석을 쓰나요?
1. **나중에 봤을 때** 코드가 뭐하는 건지 기억하려고
2. **다른 사람이 봤을 때** 이해하기 쉽게
3. **임시로 코드를 꺼둘 때** (실행하고 싶지 않은 코드)

### 주석 종류

#### 1️⃣ 한 줄 주석: `#`
```python
# 이건 주석이에요. 실행되지 않아요.
print("이건 실행돼요!")  # 이 부분도 주석
```

#### 2️⃣ 여러 줄 주석: `"""` 또는 `'''`
```python
"""
이렇게 쓰면
여러 줄을
한 번에 주석으로
만들 수 있어요!
"""
print("실행!")
```

````python
# 👉 Practice: 주석 연습
# 아래 코드를 실행하고, 주석과 코드의 차이를 확인해보세요!

# 이것은 주석입니다. 실행되지 않아요.
print("이것은 실행됩니다!")

# print("이 줄은 주석 처리되어 실행되지 않아요")
print("이것도 실행됩니다!")

"""
여러 줄 주석도
가능해요!
이 부분은 모두 무시됩니다.
"""
print("마지막 출력!")
````
