## 💡 정수(int)와 실수(float)

Python에는 두 가지 숫자 타입이 있어요!

### 1️⃣ 정수 (int, Integer)
**소수점이 없는 숫자**

- 개수를 셀 때: 학생 수, 제품 개수
- 나이, 년도
- 순서, 점수

```python
count = 10
year = 2026
score = 95
```

### 2️⃣ 실수 (float, Floating Point)
**소수점이 있는 숫자**

- 측정값: 키, 몸무게, 온도
- 비율, 평균
- 원주율 같은 수학 상수

```python
pi = 3.14
temperature = 36.5
average = 85.7
```

### 💡 큰 숫자 읽기 쉽게 쓰기
Python에서는 큰 숫자를 **언더스코어(_)**로 구분할 수 있어요!

```python
population = 51_000_000  # 5100만 (읽기 쉽죠?)
print(population)         # 51000000 (실제로는 언더스코어 없이 출력)
```

언더스코어는 보기 편하게 하려고 쓰는 거고, Python이 계산할 때는 무시해요.

---

````python
# 정수와 실수 사용하기

# 정수 (int)
count = 10
year = 2026

# 실수 (float)
pi = 3.14
temperature = 36.5

# 큰 숫자는 언더스코어로 구분 가능
population = 51_000_000  # 5100만
print(population)  # 51000000
````

## 🧮 산술 연산자 (Arithmetic Operators)

Python으로 계산할 수 있어요!

### 기본 연산자
| 연산자 | 의미 | 예시 | 결과 |
|--------|------|------|------|
| `+` | 더하기 | `10 + 3` | `13` |
| `-` | 빼기 | `10 - 3` | `7` |
| `*` | 곱하기 | `10 * 3` | `30` |
| `/` | 나누기 | `10 / 3` | `3.333...` |

### 특별한 연산자
| 연산자 | 의미 | 예시 | 결과 |
|--------|------|------|------|
| `//` | 몫 (정수 나누기) | `10 // 3` | `3` |
| `%` | 나머지 | `10 % 3` | `1` |
| `**` | 거듭제곱 | `10 ** 3` | `1000` |

### 💡 왜 세 가지 나눗셈이 있나요?

#### `/` (나누기)
- **실수** 결과가 나와요
- 정확한 값을 원할 때

#### `//` (몫)
- **정수** 결과만 (소수점 버림)
- "17개를 5명에게 나눠주면 한 명당 몇 개?" → 3개

#### `%` (나머지)
- **나머지**만 구해요
- "17개를 5명에게 나눠주고 남은 개수는?" → 2개
- 짝수/홀수 판별에 유용!

---

````python
# 산술 연산자 사용하기

# 기본 연산자
print(10 + 3)   # 13 (더하기)
print(10 - 3)   # 7  (빼기)
print(10 * 3)   # 30 (곱하기)
print(10 / 3)   # 3.333... (나누기)

# 특별한 연산자
print(10 // 3)  # 3  (몫)
print(10 % 3)   # 1  (나머지)
print(10 ** 3)  # 1000 (거듭제곱)
````

### 👉 연습 1: 연산 연습

````python
# 간단한 계산기

a = 100
b = 30

print("더하기:", a + b)
print("빼기:", a - b)
print("곱하기:", a * b)
print("나누기:", a / b)
print("몫:", a // b)
print("나머지:", a % b)
print("거듭제곱:", a ** b)
````

### 👉 연습 2: 나눗셈의 종류

````python
# 나눗셈 비교

number = 17
divisor = 5

# 일반 나누기 (/)
result1 = number / divisor
print("17 / 5 =", result1)  # 3.4

# 몫만 구하기 (//)
result2 = number // divisor
print("17 // 5 =", result2)  # 3

# 나머지만 구하기 (%)
result3 = number % divisor
print("17 % 5 =", result3)  # 2
````

## ⚡ 복합 연산자 (Compound Operators)

변수의 값을 **변경**할 때 더 짧게 쓸 수 있어요!

### 기본 방식 vs 복합 연산자

#### 기본 방식
```python
score = 80
score = score + 10  # score를 10 증가
print(score)  # 90
```

#### 복합 연산자 (같은 결과!)
```python
score = 80
score += 10  # score = score + 10과 같음!
print(score)  # 90
```

### 모든 복합 연산자
| 복합 연산자 | 의미 | 예시 | 동일한 표현 |
|-------------|------|------|-------------|
| `+=` | 더한 값을 저장 | `x += 5` | `x = x + 5` |
| `-=` | 뺀 값을 저장 | `x -= 3` | `x = x - 3` |
| `*=` | 곱한 값을 저장 | `x *= 2` | `x = x * 2` |
| `/=` | 나눈 값을 저장 | `x /= 4` | `x = x / 4` |
| `//=` | 나눈 몫을 저장 | `x //= 7` | `x = x // 7` |
| `%=` | 나눈 나머지를 저장 | `x %= 2` | `x = x % 2` |
| `**=` | 거듭제곱한 값을 저장 | `x **= 6` | `x = x ** 6` |

### 💡 언제 쓰나요?
- 점수를 누적할 때: `total_score += score`
- 카운터를 증가시킬 때: `count += 1`
- 값을 변경할 때: `price *= 1.1` (10% 인상)

---

````python
# 복합 연산자 사용하기

# 기본 방식 vs 복합 연산자
score = 80

# 기본 방식
score = score + 10
print(score)  # 90

# 복합 연산자 (같은 결과!)
score = 80
score += 10
print(score)  # 90
````

````python
##################################################
# 실습 1: 숫자 연산
##################################################

# 실습 1-1: 두 숫자의 평균 구하기
num1 = 80
num2 = 90
average = (num1 + num2) / 2
print(f"평균: {average}")
# 출력: 평균: 85.0


# 실습 1-2: 짝수 판별하기
number = 17
remainder = number % 2  # 2로 나눈 나머지
print(f"{number}를 2로 나눈 나머지: {remainder}")
print(f"{number}는 짝수인가? {remainder == 0}")
# 출력: 17를 2로 나눈 나머지: 1
# 출력: 17는 짝수인가? False

# 더 간결한 방법
is_even = number % 2 == 0
print(f"{number}는 짝수: {is_even}")


# 실습 1-3: 온도 변환
celsius = 25
fahrenheit = celsius * 9/5 + 32
print(f"섭씨 {celsius}도 = 화씨 {fahrenheit}도")
# 출력: 섭씨 25도 = 화씨 77.0도


# 실습 1-4: 거스름돈 계산
price = 7800
payment = 10000

# 거스름돈 계산
change = payment - price
print(f"거스름돈: {change}원")

# 천원짜리 개수 계산
thousand_bills = change // 1000  # 몫만 구하기
remaining = change % 1000        # 나머지
print(f"천원짜리: {thousand_bills}장")
print(f"나머지: {remaining}원")

# 출력:
# 거스름돈: 2200원
# 천원짜리: 2장
# 나머지: 200원
````
