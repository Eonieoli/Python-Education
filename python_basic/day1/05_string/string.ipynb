## ➕ 문자열 연산

문자열도 **더하고 곱할** 수 있어요!

### 문자열 더하기 (`+`)
두 문자열을 **연결**해요

```python
"Hello" + "World"  # "HelloWorld"
```

⚠️ 공백은 자동으로 안 들어가요! 직접 넣어야 해요.

### 문자열 곱하기 (`*`)
문자열을 **반복**해요

```python
"안녕" * 3  # "안녕안녕안녕"
"=" * 10    # "=========="
```

💡 **활용 예시:**
- 구분선 만들기: `"-" * 50`
- 패턴 만들기: `"*-" * 5` → `"*-*-*-*-*-"`

---

````python
##################################################
# 문자열 연산
##################################################

# 문자열 더하기 (+)
last_name = "김"
first_name = "철수"
full_name = last_name + first_name
print(full_name)  # 김철수

# 문자열 곱하기 (*)
star = "*"
line = star * 10
print(line)  # **********

# 예제
greeting = "안녕" * 3
print(greeting)  # 안녕안녕안녕
````

## 📏 문자열 길이 - len()

**len()** 함수로 문자열의 길이(글자 수)를 알 수 있어요!

```python

```

### 특징


💡 **언제 쓰나요?**

- 입력값 검사
- 문자 제한 확인

---

````python
##################################################
# 문자열 길이
##################################################

# len() 함수로 길이 구하기
text = "Hello"
length = len(text)
print(length)  # 5

# 한글도 똑같이
name = "김철수"
print(len(name))  # 3

# 공백도 포함
message = "Hello World"
print(len(message))  # 11 (공백도 센다!)
````

## 🔢 문자열 인덱싱 (Indexing)

문자열의 **특정 위치의 글자**를 가져올 수 있어요!

### 인덱스란?
문자열의 각 글자는 **번호(인덱스)**를 가지고 있어요.

```
  P   y   t   h   o   n
  0   1   2   3   4   5     (양수 인덱스)
 -6  -5  -4  -3  -2  -1     (음수 인덱스)
```

### 양수 인덱스 (앞에서부터)
- **0부터 시작**해요! (첫 글자가 0)

```python
text = "Python"
text[0]  # "P" (첫 글자)
text[1]  # "y" (두 번째)
text[5]  # "n" (여섯 번째)
```

### 음수 인덱스 (뒤에서부터)
- **-1부터 시작**해요! (마지막 글자가 -1)

```python
text = "Python"
text[-1]  # "n" (맨 뒤)
text[-2]  # "o" (뒤에서 두 번째)
text[-6]  # "P" (맨 앞)
```

💡 **언제 쓰나요?**
- 첫 글자 확인: `name[0]`
- 마지막 글자 확인: `text[-1]`

---

````python
##################################################
# 문자열 인덱싱
##################################################

# 문자열 인덱싱
text = "Python"

# 양수 인덱스
print(text[0])  # P
print(text[1])  # y
print(text[5])  # n

# 음수 인덱스
print(text[-1])  # n (맨 뒤)
print(text[-2])  # o (뒤에서 두 번째)
print(text[-6])  # P (맨 앞)
````

## ✂️ 문자열 슬라이싱 (Slicing)

문자열의 **일부분**을 잘라낼 수 있어요!

### 기본 형태
```python
문자열[시작:끝]
```

- `시작`: 포함 ✅
- `끝`: **포함 안 함** ❌

```
  P   y   t   h   o   n
  0   1   2   3   4   5   6
      ↑           ↑
   [1:5]는 "ytho" (1,2,3,4만 포함, 5는 제외!)
```

### 예시
```python
text = "Python"
text[0:6]   # "Python" (0~5까지)
text[0:3]   # "Pyt"
text[2:5]   # "tho"
```

### 생략 가능!
```python
text[:3]    # "Pyt" (처음~3 전까지)
text[3:]    # "hon" (3~끝까지)
text[:]     # "Python" (전체)
```

### 음수 인덱스도 가능!
```python
text = "Python Programming"
text[-11:]   # "Programming" (뒤에서 11번째부터 끝까지)
text[:-12]   # "Python" (처음부터 뒤에서 12번째 전까지)
```

💡 **언제 쓰나요?**
- 이메일에서 아이디 추출: `email[:email.find("@")]`
- 파일 확장자: `filename[-3:]`
- 날짜 분리: `date[0:4]` (년도)

---

````python
##################################################
# 문자열 슬라이싱
##################################################

# 문자열 슬라이싱
text = "Python Programming"

# 기본 슬라이싱
print(text[0:6])    # Python
print(text[7:18])   # Programming

# 시작/끝 생략
print(text[:6])     # Python (처음부터 6번째 전까지)
print(text[7:])     # Programming (7번째부터 끝까지)
print(text[:])      # Python Programming (전체)

# 음수 인덱스
print(text[-11:])   # Programming
print(text[:-12])   # Python
````

## 🎨 f-string으로 문자열 만들기

**f-string**은 변수를 문자열에 **쉽게 넣는** 방법이에요!

### 기본 형태
```python
f"문장 {변수} 문장"
```

- 문자열 앞에 `f` 붙이기
- 중괄호 `{}` 안에 변수 넣기

### 왜 좋나요?
#### Before (불편함 😵)
```python
name = "김철수"
age = 25
intro = "안녕하세요. 저는 " + name + "이고 " + str(age) + "살입니다."
```

#### After (편함 😊)
```python
intro = f"안녕하세요. 저는 {name}이고 {age}살입니다."
```

### 표현식도 가능!
중괄호 안에서 **계산**도 할 수 있어요!

```python
price = 10000
quantity = 3
print(f"총 금액: {price * quantity}원")  # 총 금액: 30000원
```

### 소수점 포맷팅
```python
pi = 3.141592
print(f"파이: {pi:.2f}")  # 파이: 3.14 (소수점 2자리)
```

- `:.2f` → 소수점 2자리까지

💡 **FastAPI에서 필수!**
- JSON 응답 만들 때
- 로그 메시지 만들 때
- 계속 쓰게 될 거예요!

---

````python
##################################################
# 문자열 포매팅 (f-string)
##################################################

# f-string 사용하기
name = "김철수"
age = 25
city = "서울"

# 기본 사용
intro = f"안녕하세요. 저는 {name}입니다."
print(intro)

# 여러 변수
profile = f"이름: {name}, 나이: {age}살, 거주지: {city}"
print(profile)

# 표현식도 가능
price = 10000
quantity = 3
total = f"총 금액: {price * quantity}원"
print(total)  # 총 금액: 30000원

# 소수점 포맷팅
pi = 3.141592
print(f"파이: {pi:.2f}")  # 파이: 3.14 (소수점 2자리)
````

## 🔧 문자열 메서드 (1) - 대소문자 변환

문자열에는 **메서드**(함수)가 많이 있어요!

### 메서드 사용법
```python
문자열.메서드()
```

점(`.`)을 찍고 메서드 이름을 쓰면 돼요!

### 대소문자 변환 메서드

| 메서드 | 기능 | 예시 |
|--------|------|------|
| `.upper()` | 모두 대문자로 | `"hello"` → `"HELLO"` |
| `.lower()` | 모두 소문자로 | `"HELLO"` → `"hello"` |
| `.capitalize()` | 첫 글자만 대문자 | `"hello world"` → `"Hello world"` |
| `.title()` | 각 단어 첫 글자 대문자 | `"hello world"` → `"Hello World"` |
| `.swapcase()` | 대소문자 바꾸기 | `"Hello"` → `"hELLO"` |

💡 **언제 쓰나요?**
- 대소문자 통일: `email.lower()`
- 제목 포맷팅: `title.title()`
- 비교할 때: `input().upper() == "YES"`

---

````python
##################################################
# 문자열 메서드 (대소문자 변환)
##################################################

# 문자열 메서드 (대소문자)
text = "Hello World"

# 대문자로
print(text.upper())         # HELLO WORLD

# 소문자로
print(text.lower())         # hello world

# 첫 글자만 대문자
print(text.capitalize())    # Hello world

# 제목 형식 (각 단어 첫 글자 대문자)
print(text.title())         # Hello World

# 대소문자 바꾸기
print(text.swapcase())      # hELLO wORLD
````

## 🔧 문자열 메서드 (2) - 공백 제거 & 분리/결합

### 공백 제거 메서드

| 메서드 | 기능 |
|--------|------|
| `.strip()` | 양쪽 공백 제거 |
| `.lstrip()` | 왼쪽 공백 제거 |
| `.rstrip()` | 오른쪽 공백 제거 |

```python
text = "  Hello  "
text.strip()   # "Hello"
text.lstrip()  # "Hello  "
text.rstrip()  # "  Hello"
```

### 분리: `.split()`
문자열을 **리스트**로 나눠요!

```python
sentence = "Python is awesome"
words = sentence.split()  # ['Python', 'is', 'awesome']

email = "user@example.com"
parts = email.split("@")  # ['user', 'example.com']
```

- 기본: 공백으로 분리
- 특정 문자로 분리 가능: `.split("@")`

### 결합: `.join()`
리스트를 **문자열**로 합쳐요!

```python
words = ["Python", "is", "fun"]
sentence = " ".join(words)  # "Python is fun"
```

- `" "` → 공백으로 연결
- `"-"` → 하이픈으로 연결
- `""` → 붙여쓰기

💡 **실전 활용:**
- 사용자 입력 정제: `input().strip()`
- CSV 파싱: `line.split(",")`
- URL 생성: `"-".join(words)`

---

````python
##################################################
# 문자열 메서드 (공백/분리)
##################################################

# 공백 제거
text = "  Hello World  "
print(text.strip())   # "Hello World" (양쪽 공백 제거)
print(text.lstrip())  # "Hello World  " (왼쪽만)
print(text.rstrip())  # "  Hello World" (오른쪽만)

# 문자열 분리 (split)
sentence = "Python is awesome"
words = sentence.split()  # 공백 기준 분리
print(words)  # ['Python', 'is', 'awesome']

email = "user@example.com"
parts = email.split("@")  # @ 기준 분리
print(parts)  # ['user', 'example.com']

# 문자열 결합 (join)
words = ["Python", "is", "fun"]
sentence = " ".join(words)
print(sentence)  # "Python is fun"
````

## 🔧 문자열 메서드 (3) - 검색 & 치환

### 찾기: `.find()`
특정 문자열의 **위치(인덱스)**를 찾아요!

```python
text = "Hello World"
text.find("World")   # 6 (시작 위치)
text.find("Python")  # -1 (없으면 -1)
```

### 포함 여부: `in`
문자열이 **포함되어 있는지** 확인해요!

```python
"World" in "Hello World"   # True
"Python" in "Hello World"  # False
```

- 결과: `True` 또는 `False`

### 치환: `.replace()`
특정 문자열을 **다른 문자열로 바꿔요**!

```python
message = "Hello World"
new_message = message.replace("World", "Python")
# "Hello Python"
```

- 모든 일치하는 부분을 바꿔요

```python
text = "banana"
text.replace("a", "o")  # "bonono" (모든 a가 o로)
```

💡 **실전 활용:**
- 검색: `if "error" in log:`
- 정제: `text.replace(" ", "-")`

````python
##################################################
# 실습 2: 문자열 인덱싱/슬라이싱
##################################################

# 실습 2-1: 주민등록번호 나누기
jumin = "901015-1234567"

# 방법 1: 슬라이싱으로 직접 추출
birth_date = jumin[0:6]      # 또는 jumin[:6]
back_number = jumin[7:14]     # 또는 jumin[7:]
print(f"생년월일: {birth_date}")
print(f"뒷자리: {back_number}")

# 방법 2: split 메서드 사용 (더 간편!)
parts = jumin.split("-")
print(f"생년월일: {parts[0]}")
print(f"뒷자리: {parts[1]}")

# 출력:
# 생년월일: 901015
# 뒷자리: 1234567


# 실습 2-2: 파일명과 확장자 분리
filename = "document.pdf"

# 방법 1: 슬라이싱 (점의 위치가 6)
name_part = filename[0:8]     # 또는 filename[:8]
extension = filename[9:12]     # 또는 filename[9:]
print(f"파일명: {name_part}")
print(f"확장자: {extension}")

# 방법 2: split 메서드 사용 (추천!)
parts = filename.split(".")
print(f"파일명: {parts[0]}")
print(f"확장자: {parts[1]}")

# 출력:
# 파일명: document
# 확장자: pdf


# 실습 2-3: 전화번호 포맷팅
phone = "01012345678"

# 슬라이싱으로 나누기
part1 = phone[0:3]      # "010"
part2 = phone[3:7]      # "1234"
part3 = phone[7:11]     # "5678"

# - 로 연결
formatted = part1 + "-" + part2 + "-" + part3
print(formatted)

# f-string으로 더 깔끔하게
formatted = f"{phone[0:3]}-{phone[3:7]}-{phone[7:11]}"
print(formatted)

# 출력: 010-1234-5678


# 실습 2-4: 이름 첫 글자
name = "김철수"

# 슬라이싱
last_name = name[0]       # 또는 name[0:1]
first_name = name[1:3]    # 또는 name[1:]

print(f"성: {last_name}")
print(f"이름: {first_name}")

# 출력:
# 성: 김
# 이름: 철수
````

````python
##################################################
# 실습 3: 문자열 처리
##################################################

# 실습 3-1: 이메일 정제
email = "  USER@EXAMPLE.COM  "

# 단계별 처리
cleaned = email.strip()        # 공백 제거
cleaned = cleaned.lower()      # 소문자 변환
print(cleaned)

# 한 줄로 (메서드 체이닝)
cleaned = email.strip().lower()
print(cleaned)

# 출력: user@example.com


# 실습 3-2: CSV 데이터 파싱
data = "김철수,25,서울,개발자"

# split으로 분리
items = data.split(",")

# 각 항목 출력
print(f"이름: {items[0]}")
print(f"나이: {items[1]}")
print(f"지역: {items[2]}")
print(f"직업: {items[3]}")

# 출력:
# 이름: 김철수
# 나이: 25
# 지역: 서울
# 직업: 개발자


# 실습 3-3: 비밀번호 검증
password = "mypassword123"

# 길이 확인
length_ok = len(password) >= 8
print(f"길이 8자 이상: {length_ok}")

# 숫자 포함 확인 (여러 방법)
has_number = "123" in password
print(f"숫자 포함: {has_number}")

# 더 정확한 방법 (어떤 숫자든 포함하는지)
has_digit = any(char.isdigit() for char in password)
print(f"숫자 포함 (정확): {has_digit}")

# 출력:
# 길이 8자 이상: True
# 숫자 포함: True


# 실습 3-4: URL 슬러그 생성
title = "Python Programming Tutorial"

# 단계별 처리
slug = title.lower()              # 소문자로
slug = slug.replace(" ", "-")     # 공백을 -로
print(slug)

# 한 줄로
slug = title.lower().replace(" ", "-")
print(slug)

# 출력: python-programming-tutorial
````
