**💡 집합(Set)이란?**
- 중괄호 `{}`로 만들어요
- **순서가 없어요!**
- **중복을 허용하지 않아요!**
- 리스트의 중복 제거할 때 유용!

---

## 📝 집합의 특징

### 집합 vs 리스트 vs 튜플

| 특징 | 리스트 | 튜플 | 집합 |
|------|--------|------|------|
| 괄호 | `[]` | `()` | `{}` |
| 순서 | ✅ 있음 | ✅ 있음 | ❌ 없음 |
| 중복 | ✅ 가능 | ✅ 가능 | ❌ 불가능 |
| 수정 | ✅ 가능 | ❌ 불가능 | ✅ 가능 |
| 인덱싱 | ✅ 가능 | ✅ 가능 | ❌ 불가능 |

### 언제 사용하나요?
- **중복 제거**: 리스트에서 중복된 값 제거할 때
- **빠른 검색**: `in` 연산자가 매우 빠름!
- **집합 연산**: 합집합, 교집합 등

---

## 🔨 집합 만들기

### 집합 생성
```python
fruits = {"apple", "banana", "orange"}
```

### ⚠️ 주의사항

**빈 집합**
```python
empty_set = set()  # ✅ 빈 집합
empty_dict = {}    # ❌ 빈 딕셔너리!
```

**순서가 없어요!**
- 출력할 때마다 순서가 바뀔 수 있어요
- 인덱싱 불가: `fruits[0]` ❌

**중복 자동 제거**
- 같은 값을 여러 번 넣어도 하나만 남아요

---

````python
##################################################
# 집합으로 만들기
##################################################

# 집합 생성 - 중괄호 {} 사용
fruits = {"apple", "banana", "orange"}
print(fruits)  # {'banana', 'apple', 'orange'}
# 순서가 뒤죽박죽! 순서 없음!

# 빈 집합 생성
empty_set = set()       # empty_set = {} 는 빈 딕셔너리 생성
print(empty_set)

# 중복된 값을 넣어도 자동으로 하나만 남음
colors = {"red", "blue", "red", "green", "blue"}
print(colors)  # {'blue', 'green', 'red'}

# in 연산자 - 매우 빠름!
print("apple" in fruits)  # True
print("grape" in fruits)  # False
````

## ✨ 집합으로 중복 제거하기

집합의 **가장 실용적인 사용법**!

### 중복 제거 과정
```python
리스트 → set() → 리스트
[1,2,2,3] → {1,2,3} → [1,2,3]
```

### 활용 예시
- 중복된 이메일 주소 제거
- 중복된 태그 제거
- 유일한 값만 추출

---

````python
##################################################
# 집합으로 중복 제거하기
##################################################

# 중복이 있는 리스트
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
print(numbers)  # [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

# set으로 변환하면 중복 제거!
unique_numbers = set(numbers)
print(unique_numbers)  # {1, 2, 3, 4}

# 다시 리스트로 변환
result = list(unique_numbers)
print(result)  # [1, 2, 3, 4]
````

## 💡 핵심 정리

### ✅ 집합의 3가지 특징
1. **순서 없음**: 인덱싱 불가
2. **중복 불가**: 자동으로 중복 제거
3. **빠른 검색**: `in` 연산이 빠름

### 📌 기억하세요!
- 빈 집합: `set()` (✅) vs `{}` (❌ 딕셔너리)
- 중복 제거: `list(set(리스트))`
- 순서가 중요하면 리스트, 중복 제거가 필요하면 집합!
