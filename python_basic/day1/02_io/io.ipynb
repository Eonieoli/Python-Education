## 💡 print() 함수란?

**print()**는 Python에서 가장 많이 쓰는 함수예요!

### 하는 일
- 괄호 `()` 안의 내용을 **화면에 출력**해요
- 디버깅할 때, 값을 확인할 때, 사용자에게 메시지를 보여줄 때 사용해요

### 기본 형태
```python
print("출력할 내용")
```

자, 이제 직접 해볼까요? 👇

---

````python
# 👉 Practice 1: print() 기본 사용
# 아래 코드를 실행해보세요! (Shift + Enter)

print("Hello World")
print("안녕하세요!")
````

````python
# 👉 Practice 2: 여러 개 출력 (쉼표로 구분)
# 쉼표(,)로 구분하면 여러 값을 한 번에 출력할 수 있어요

print("이름:", "김철수")
print("나이:", 25, "살")

# 💡 쉼표로 구분하면 자동으로 공백이 들어가요!
````

````python
# 👉 Practice 3: 계산 결과 출력
# print() 안에서 바로 계산할 수 있어요!

print(10 + 20)        # 30
print(5 * 3)          # 15
print(100 - 30)       # 70
print(10 / 2)         # 5.0

# 💡 계산 결과가 바로 출력돼요!
````

````python
# 👉 Practice 4: 변수 출력
# 변수에 저장된 값도 출력할 수 있어요

name = "이영희"
age = 30

print(name)           # 이영희
print(name, age)      # 이영희 30
````

## 💡 f-string으로 문장 만들기

**f-string**은 문자열 안에 변수를 쉽게 넣는 방법이에요!

### 기본 형태
```python
f"문장 {변수} 문장"
```

- 문자열 앞에 `f` 붙이기
- 중괄호 `{}` 안에 변수 이름 넣기

### 왜 좋나요?
- 문장을 자연스럽게 만들 수 있어요
- 나중에 FastAPI에서 엄청 많이 쓸 거예요!

---

````python
# 👉 Practice 5: f-string 사용하기

name = "이영희"
age = 30

# 일반 방법
print(name, "님은", age, "살입니다")

# f-string 방법 (더 자연스럽죠?)
print(f"{name}님은 {age}살입니다")

# 더 복잡한 예제
height = 170
weight = 65
print(f"{name}님의 키는 {height}cm이고, 몸무게는 {weight}kg입니다.")

# 계산 결과도 넣을 수 있어요!
print(f"내년엔 {age + 1}살이네요!")
````

## 💡 input() 함수란?

**input()**은 사용자로부터 **입력을 받는** 함수예요!

### 하는 일
- 프로그램이 실행되면 입력을 기다려요
- 사용자가 뭔가 입력하고 Enter를 누르면 그 값을 받아와요

### 기본 형태
\```python
변수 = input("메시지")
\```

### 언제 쓰나요?
- 사용자 이름 물어볼 때
- 나이 물어볼 때
- 선택지 고르게 할 때

⚠️ **중요**: input()으로 받은 값은 **항상 문자열(str)**이에요!

---

````python
# 👉 Practice 6: input() 기본 사용
# 아래 코드를 실행하고, 이름을 입력해보세요!

name = input("이름을 입력하세요: ")
print(f"안녕하세요, {name}님!")

# 💡 실행하면 입력 칸이 나와요
# 이름을 입력하고 Enter를 누르세요!
````

## ⚠️ 주의! input()은 항상 문자열을 돌려줘요

### 함정 예시
\```python
age = input("나이를 입력하세요: ")  # "25" 입력
print(type(age))  # <class 'str'> - 문자열!
\```

숫자를 입력해도 **문자열**로 저장돼요!

### 왜 문제가 될까요?
\```python
age = input("나이: ")  # "25" 입력
next_year = age + 1    # ❌ 에러! 문자열 + 숫자는 안 돼요
\```

### 해결 방법: int()로 변환!
\```python
age = input("나이: ")      # "25" (문자열)
age = int(age)             # 25 (숫자)
next_year = age + 1        # ✅ 26
\```

---

````python
# 👉 Practice 7: input()의 타입 확인
# 숫자를 입력해도 문자열로 저장됩니다!

age = input("나이를 입력하세요: ")
print(f"입력한 값: {age}")
print(f"타입: {type(age)}")  # <class 'str'>

# 💡 숫자처럼 보이지만 사실은 문자열!
````

````python
# 👉 Practice 8: int()로 숫자로 변환하기

# 방법 1: 두 줄로
age = input("나이를 입력하세요: ")
age = int(age)  # 문자열을 숫자로 변환!
next_year = age + 1
print(f"내년엔 {next_year}살이네요!")

print("---")

# 방법 2: 한 줄로 (더 간결!)
age = int(input("나이를 입력하세요: "))
result = age * 2
print(f"두 배는 {result}살!")
````

## 💡 int() 변환 정리

### 언제 int()를 쓰나요?
✅ 입력받은 값으로 **계산**을 해야 할 때  
✅ 나이, 점수, 개수 같은 **숫자** 데이터일 때

### 방법 비교
```python
# 방법 1: 나눠서 (초보자 추천)
age = input("나이: ")
age = int(age)

# 방법 2: 한 줄로 (익숙해지면)
age = int(input("나이: "))
```

둘 다 결과는 같아요! 편한 방법으로 쓰세요 😊
