## 📝 모듈이란?

## 🔨 import 기본 (4가지 방법)

### 방법 1: 모듈 전체 import
```python
import random
random.randint(1, 10)  # 모듈명.함수명
```

### 방법 2: 특정 기능만 import
```python
from random import randint, choice
randint(1, 100)  # 바로 사용
```

### 방법 3: 별칭 사용 (alias)
```python
import random as rd
rd.randint(1, 6)
```

### 방법 4: 전체 import (비권장 ❌)
```python
from random import *
# 모든 기능을 가져옴 (이름 충돌 위험)
```

**💡 추천 방법:**
- 방법 1 또는 방법 2 사용
- FastAPI: `from fastapi import FastAPI`

---

````python
##################################################
# import 기본
##################################################

# 방법 1: 모듈 전체 import
import random

# 모듈명.함수명으로 사용
number = random.randint(1, 10)
print(f"랜덤 숫자: {number}")

choice = random.choice(["사과", "바나나", "오렌지"])
print(f"선택: {choice}")

# 방법 2: 특정 기능만 import
from random import randint, choice

# 모듈명 없이 바로 사용
number = randint(1, 100)
fruit = choice(["딸기", "수박", "포도"])

# 방법 3: 별칭 사용
import random as rd

num = rd.randint(1, 6)  # 주사위
print(f"주사위: {num}")

# 방법 4: 전체 import (비권장)
from random import *
# 모든 기능을 가져옴 (충돌 위험)
````

## 🎲 random 모듈 - 랜덤 기능

### 주요 함수

| 함수 | 설명 | 예시 |
|------|------|------|
| `randint(a, b)` | a~b 사이 정수 | `randint(1, 6)` → 주사위 |
| `choice(리스트)` | 리스트에서 랜덤 선택 | `choice(["A", "B"])` |
| `sample(리스트, n)` | n개 선택 (중복 없이) | `sample(range(1, 46), 6)` → 로또 |
| `shuffle(리스트)` | 리스트 섞기 | `shuffle(cards)` |
| `random()` | 0.0~1.0 실수 | `random()` → 0.7234... |

**💡 활용 예시:**
- 게임 개발 (주사위, 카드)
- 랜덤 추첨
- 테스트 데이터 생성

---

### 👉 연습 1: random 모듈 활용

````python
##################################################
# 연습 1: random 모듈 활용
##################################################

import random

# randint: 정수 난수
dice = random.randint(1, 6)
print(f"주사위: {dice}")

# choice: 리스트에서 랜덤 선택
menu = ["김치찌개", "된장찌개", "순두부"]
today_menu = random.choice(menu)
print(f"오늘의 메뉴: {today_menu}")

# sample: 여러 개 선택 (중복 없이)
numbers = list(range(1, 46))  # 1~45
lotto = random.sample(numbers, 6)
lotto.sort()
print(f"로또 번호: {lotto}")

# shuffle: 리스트 섞기
cards = ["◆A", "◆2", "♥A", "♥2", "♣A", "♣2", "♠A", "♠2"]
random.shuffle(cards)
print(f"섞인 카드: {cards}")

# random: 0.0~1.0 실수 난수
probability = random.random()
print(f"확률: {probability:.2f}")

if probability > 0.5:
    print("성공!")
else:
    print("실패!")
````

## 📅 datetime 모듈 - 날짜/시간 다루기

### 주요 기능

**현재 날짜/시간:**
```python
from datetime import datetime
now = datetime.now()  # 현재 시간
```

**날짜 포맷:**
```python
now.strftime('%Y-%m-%d')  # 2024-01-16
now.strftime('%Y년 %m월 %d일')  # 2024년 01월 16일
```

**날짜 계산:**
```python
from datetime import timedelta
tomorrow = now + timedelta(days=1)
next_week = now + timedelta(weeks=1)
```

**시간 차이:**
```python
diff = future_date - now
print(diff.days)  # 며칠 차이
```

**💡 활용 예시:**
- 로그 타임스탬프
- D-day 계산
- 예약 시스템

---

### 👉 연습 2: datetime 모듈 활용

````python
##################################################
# 연습 2: datetime 모듈 활용
##################################################

from datetime import datetime, timedelta

# 현재 시간
now = datetime.now()
print(f"현재: {now}")
print(f"포맷: {now.strftime('%Y-%m-%d %H:%M:%S')}")

# 날짜 부분만
print(f"년: {now.year}")
print(f"월: {now.month}")
print(f"일: {now.day}")
print(f"시: {now.hour}")

# 날짜 계산
tomorrow = now + timedelta(days=1)
print(f"내일: {tomorrow.strftime('%Y-%m-%d')}")

next_week = now + timedelta(weeks=1)
print(f"다음주: {next_week.strftime('%Y-%m-%d')}")

# 시간 차이
christmas = datetime(2026, 12, 25)
diff = christmas - now
print(f"크리스마스까지: {diff.days}일")

# 특정 날짜 생성
meeting = datetime(2026, 1, 20, 14, 30)
print(f"회의: {meeting.strftime('%Y년 %m월 %d일 %H시 %M분')}")
````

## 📦 json 모듈 - 데이터 변환

### JSON이란?
**JavaScript Object Notation**
- 데이터를 주고받을 때 사용하는 **텍스트 형식**
- Python 딕셔너리와 **매우 유사**
- **FastAPI의 핵심**: API는 JSON으로 통신!

### 주요 함수

**Python → JSON (직렬화):**
```python
import json
user = {"name": "김철수", "age": 25}
json_str = json.dumps(user, ensure_ascii=False)
# '{"name": "김철수", "age": 25}'
```

**JSON → Python (역직렬화):**
```python
json_data = '{"name": "이영희", "age": 30}'
user_dict = json.loads(json_data)
# {'name': '이영희', 'age': 30}
```

**💡 FastAPI에서:**
```python
@app.post("/users")
def create_user(user: dict):
    # FastAPI가 자동으로 JSON → dict 변환!
    return user  # 자동으로 dict → JSON 변환!
```

---

### 👉 연습 3: json 모듈 활용

````python
##################################################
# 연습 3: json 모듈 활용
##################################################

import json

# Python 딕셔너리 → JSON 문자열
user = {
    "name": "김철수",
    "age": 25,
    "email": "kim@test.com"
}

json_str = json.dumps(user, ensure_ascii=False)
print(json_str)
# {"name": "김철수", "age": 25, "email": "kim@test.com"}
print(type(json_str))  # <class 'str'>

# JSON 문자열 → Python 딕셔너리
json_data = '{"name": "이영희", "age": 30}'
user_dict = json.loads(json_data)
print(user_dict)
print(type(user_dict))  # <class 'dict'>
print(user_dict["name"])  # 이영희

# 리스트도 가능
users = [
    {"name": "김철수", "age": 25},
    {"name": "이영희", "age": 30}
]

json_str = json.dumps(users, ensure_ascii=False, indent=2)
print(json_str)

# 파일 저장/읽기 (개념만)
# with open('data.json', 'w') as f:
#     json.dump(user, f)
````

## 🆚 dumps vs dump, loads vs load

### dumps / loads (문자열)
- **dumps**: dict → JSON **문자열**
- **loads**: JSON **문자열** → dict
- "s"는 **string**을 의미

### dump / load (파일)
- **dump**: dict → JSON **파일**
- **load**: JSON **파일** → dict

**💡 FastAPI에서는 주로 dumps/loads 사용!**
