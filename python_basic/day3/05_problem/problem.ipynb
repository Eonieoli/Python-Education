## 문제 10: 성적순 정렬 (+ 동점자 처리)

**요구사항:**
- 학생들을 평균 점수 **내림차순**으로 정렬하세요
- 평균이 같으면 **이름순**(오름차순)으로 정렬
- sorted() 함수와 lambda 사용

**힌트:**
```python
# 평균 계산 함수
def get_average(student):
    return (student["korean"] + student["english"] + student["math"]) / 3

# 정렬 (평균 내림차순, 이름 오름차순)
sorted_students = sorted(
    students,
    key=lambda s: (-get_average(s), s["name"])
)

# 출력
print("=== 성적순 명단 ===")
for rank, student in enumerate(sorted_students, start=1):
    avg = get_average(student)
    print(f"{rank}등: {student['name']}님 ({avg:.1f}점)")
```

**설명:**
- `-get_average(s)`: 음수를 붙여서 내림차순
- `s["name"]`: 이름은 그대로 오름차순
- 튜플로 묶으면 첫 번째 기준으로 먼저 정렬, 같으면 두 번째 기준 적용
