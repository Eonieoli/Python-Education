## 🛡️ try-except 기본

### 예외를 안전하게 처리하기

```python
try:
    # 예외가 발생할 수 있는 코드
    위험한_코드()
except 예외타입:
    # 예외 발생 시 실행
    처리_코드()
```

### try-except의 동작
1. `try` 블록 실행
2. 예외 발생 → `except` 블록으로 이동
3. 예외 없음 → `except` 블록 건너뛰기
4. **프로그램 계속 실행!** ✅

### 예외 처리 전 vs 후
```python
# ❌ 예외 처리 없음
result = 10 / 0  # 여기서 프로그램 종료
print("실행 안 됨")

# ✅ 예외 처리 있음
try:
    result = 10 / 0
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다")
print("프로그램 계속 실행!")  # ← 실행됨!
```

---

````python
##################################################
# try-except 기본
##################################################

# 예외 처리 없이 (프로그램 종료)
# result = 10 / 0  # ❌ 여기서 프로그램 멈춤
# print("이 줄은 실행 안 됨")

# try-except 사용 (프로그램 계속 실행)
try:
    result = 10 / 0
    print("나눗셈 성공")  # 이 줄은 실행 안 됨
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다")  # 이 줄 실행!

print("프로그램 계속 실행됨")  # ✅ 이 줄도 실행됨!

# 출력:
# 0으로 나눌 수 없습니다
# 프로그램 계속 실행됨
````

````python
##################################################
# 특정 예외 처리하기
##################################################

# 여러 종류의 예외 처리
user_input = input("숫자를 입력하세요: ")

try:
    number = int(user_input)  # ValueError 가능
    result = 100 / number      # ZeroDivisionError 가능
    print(f"결과: {result}")
    
except ValueError:
    print("숫자를 입력해주세요!")
    
except ZeroDivisionError:
    print("0이 아닌 숫자를 입력해주세요!")

# 입력: "abc" → "숫자를 입력해주세요!"
# 입력: "0"   → "0이 아닌 숫자를 입력해주세요!"
# 입력: "5"   → "결과: 20.0"
````

## 💬 예외 메시지 받기 (as)

### 예외 객체를 변수로 받기

```python

    print(f"에러: {e}")  # 에러 메시지 출력
```

### 왜 필요한가?
- 어떤 에러가 발생했는지 **정확히 알 수 있음**
- 에러 메시지를 **로그에 기록** 가능
- 사용자에게 **구체적인 안내** 가능

### 실전 활용
```python
except KeyError as e:
    print(f"'{e}' 키가 존재하지 않습니다")
```

---

````python
##################################################
# 예외 메시지 받기
##################################################

# as 키워드로 예외 객체 받기
data = {"name": "김철수", "age": 25}

try:
    email = data["email"]  # KeyError 발생
    
except KeyError as e:
    print(f"에러 발생: {e}")
    print(f"'{e}' 키가 존재하지 않습니다")

# 출력:
# 에러 발생: 'email'
# 'email' 키가 존재하지 않습니다

# 다른 예시
try:
    number = int("hello")
except ValueError as e:
    print(f"변환 실패: {e}")
````

### 👉 연습 1: 기본 예외 처리

````python
##################################################
# 연습 1: 기본 예외 처리
##################################################

# 나이 입력 받기 (안전하게)
def get_age_safely(age_str):
    try:
        age = int(age_str)
        if age < 0:
            return None
        return age
    except ValueError:
        print("숫자를 입력해주세요!")
        return None

# 테스트
print(get_age_safely("25"))      # 25
print(get_age_safely("스물다섯"))  # 숫자를 입력해주세요! None
print(get_age_safely("-5"))      # None

# 딕셔너리 안전하게 접근
def get_value_safely(data, key, default=None):
    try:
        return data[key]
    except KeyError:
        print(f"'{key}' 키가 없습니다!")
        return default

user = {"name": "김철수", "age": 25}
print(get_value_safely(user, "name"))      # 김철수
print(get_value_safely(user, "email", "없음"))  # '이메일' 키가 없습니다! 없음

# 리스트 안전하게 접근
def get_item_safely(items, index):
    try:
        return items[index]
    except IndexError:
        print(f"인덱스 {index}는 범위를 벗어났습니다!")
        return None

fruits = ["apple", "banana", "orange"]
print(get_item_safely(fruits, 1))   # banana
print(get_item_safely(fruits, 10))  # 에러 메시지, None
````

## 🔄 raise + try-except 조합

### 에러를 던지고 받기

```python
# 함수에서 raise
def validate(data):
    if not data:
        raise ValueError("데이터 없음")

# 호출하는 쪽에서 try-except
try:
    validate(None)

    print(f"검증 실패: {e}")
```

### 패턴 정리
1. **함수 안에서**: 문제 발견 → `raise`로 에러 던지기
2. **함수 밖에서**: `try-except`로 에러 잡기
3. 프로그램은 계속 실행됨 ✅

**💡 FastAPI에서:** 컨트롤러에서 raise, FastAPI가 자동으로 HTTP 응답 변환!

---

````python
##################################################
# raise + try-except 조합
##################################################

# raise와 try-except를 함께 사용
def validate_age(age):
    if age < 0:
        raise ValueError("나이는 0 이상이어야 합니다.")
    return age

# 예외 처리하면서 사용
try:
    result = validate_age(-5)
    print(f"검증 성공: {result}세")
except ValueError as e:
    print(f"검증 실패: {e}")
print("프로그램 계속 실행")
````

## ✅ else - 예외가 없을 때

### try 성공 시에만 실행

```python
try:
    위험한_코드()  # 여기만 try!
except 예외타입:
    print("실패")
else:
    print("성공!")  # 예외 없을 때만
```

### else vs try 블록 안에 넣기

| 방식 | 특징 |
|------|------|
| try 안에 다 넣기 | 코드가 많아지면 혼란스러움 |
| else 사용 | **의도가 명확**, try는 짧게! ✅ |

### 왜 else가 좋은가?
1. **try 블록은 짧게** - 에러 발생 가능한 부분만
2. **의도가 명확** - 시도 vs 성공 후 처리 구분
3. **else 블록의 에러는 catch 안 됨** - 예상치 못한 에러 놓치지 않음

---

````python
##################################################
# try vs else
##################################################

# 방법 1: try 블록에 모두 넣기
try:
    result = 10 / 2
    print("계산 성공!")  # 예외 없으면 실행
    print(f"결과: {result}")
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다")

# 방법 2: else 블록 활용 (더 명확!)
try:
    result = 10 / 2  # 시도하는 부분만!
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다")
else:
    # 성공했을 때만 실행
    print("계산 성공!")
    print(f"결과: {result}")

# 왜 else가 좋은가?
# 1. 의도가 명확함 (시도 vs 성공 후 처리)
# 2. try 블록은 짧게! (에러 발생 가능한 부분만)
# 3. else 블록에서 추가 에러 발생 시 catch 안 됨
````

## 🔒 finally - 항상 실행

### 예외 여부와 상관없이 무조건 실행

```python
try:
    위험한_코드()
except 예외타입:
    print("실패")
finally:
    print("항상 실행!")  # ← 무조건!
```

### 언제 사용하나?
- **리소스 정리**: 파일 닫기, DB 연결 종료
- **로그 기록**: 성공/실패 여부 기록
- **카운터 증가**: 시도 횟수 추적

### 실행 순서
```python
try:
    코드  # 1번 실행
except:
    에러처리  # 2번 (에러 시)
else:
    성공처리  # 2번 (성공 시)
finally:
    정리  # 3번 (항상)
```

**💡 FastAPI에서:** DB 연결, 파일 처리 등에서 finally로 정리!

---

````python
##################################################
# finally
##################################################

# finally 사용 예시
def read_file(filename):
    file = None
    try:
        file = open(filename, 'r', encoding='utf-8')
        content = file.read()
        print(content)
        return content
        
    except FileNotFoundError:
        print(f"파일 {filename}을 찾을 수 없습니다")
        
    finally:
        if file:
            file.close()
            print("파일을 닫았습니다")

# 파일이 있는 경우
# read_file("test.txt")
# 출력:
# (파일 내용)
# 파일을 닫았습니다

# 파일이 없는 경우
read_file("없는파일.txt")
# 출력:
# 파일 없는파일.txt을 찾을 수 없습니다
# 파일을 닫았습니다  ← finally는 실행됨!
````

## 🎯 try-except-else-finally 완전체

### 모든 블록을 함께 사용

```python
try:
    # 1. 시도할 코드
    위험한_코드()
except 예외타입:
    # 2. 예외 발생 시
    에러_처리()
else:
    # 3. 예외 없을 때만
    성공_처리()
finally:
    # 4. 항상 실행
    정리_작업()
```

### 실행 흐름

**성공 시:**  
try → else → finally

**실패 시:**  
try → except → finally

### 완전한 예외 처리
- **try**: 위험한 코드만 짧게
- **except**: 예외별로 다르게 처리
- **else**: 성공 후 추가 로직
- **finally**: 무조건 정리

---

````python
##################################################
# try-except-else-finally
##################################################

# 완전한 예외 처리 구조
def divide_safely(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("0으로 나눌 수 없습니다!")
        return None
    except TypeError:
        print("숫자만 입력 가능합니다!")
        return None
    else:
        # 예외가 없을 때만 실행
        print("계산 성공!")
        return result
    finally:
        # 항상 실행 (에러 여부 무관)
        print("함수 종료")

print(divide_safely(10, 2))
# 계산 성공!
# 함수 종료
# 5.0

print(divide_safely(10, 0))
# 0으로 나눌 수 없습니다!
# 함수 종료
# None
````
