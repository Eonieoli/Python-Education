## 📝 None의 특징

### None이란?
- Python의 **특별한 값**
- "아무 값도 없음"을 표현
- 타입은 `NoneType`

### 언제 사용하나요?

**1. 초기값**
```python
user_name = None  # 아직 입력 안 받음
```

**2. 선택적 값**
```python
profile_image = None  # 프로필 사진 없음
```

**3. 함수 반환**
```python
# 반환값이 없는 함수는 자동으로 None 반환
```

---

````python
##################################################
# None 기본 사용 방법
##################################################

# None 할당
profile_image = None
print(profile_image)  # None
print(type(profile_image))  # <class 'NoneType'>

# None 체크하기
if profile_image is None:
    print("프로필 사진이 없습니다")
else:
    print("프로필 사진:", profile_image)

# 함수가 아무것도 반환하지 않으면 None
def greet():
    print("안녕하세요!")

result = greet()  # 안녕하세요! 출력
print(result)     # None
````

## 💡 핵심 포인트

### 1️⃣ None 체크는 `is` 사용!
```python
# ✅ 권장
if value is None:
    ...

# ❌ 비권장
if value == None:
    ...
```

### 2️⃣ 함수의 기본 반환값
- `return`이 없는 함수는 **자동으로 None 반환**
- `return`만 쓰면 None 반환

```python
def func1():
    print("hello")
    # return이 없음 → None 반환

def func2():
    print("hello")
    return  # 명시적으로 None 반환
```

### 3️⃣ None vs 0 vs 빈 문자열

| 값 | 의미 | 타입 |
|---|------|------|
| `None` | 값 없음 | `NoneType` |
| `0` | 숫자 0 | `int` |
| `""` | 빈 문자열 | `str` |
| `[]` | 빈 리스트 | `list` |

**모두 다른 값이에요!**
