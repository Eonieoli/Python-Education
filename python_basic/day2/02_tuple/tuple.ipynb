- 괄호 `()`로 만들어요
- 좌표, RGB 색상처럼 **변하지 않는 데이터**에 사용!

---

## 📝 튜플의 특징

### 튜플 vs 리스트

| 특징 | 리스트 | 튜플 |
|------|--------|------|
| 괄호 | `[]` | `()` |
| 수정 | ✅ 가능 | ❌ 불가능 |
| 속도 | 보통 | 빠름 |
| 용도 | 변경되는 데이터 | 고정된 데이터 |

### 언제 사용하나요?
- 좌표: `(x, y)`
- RGB 색상: `(255, 0, 0)`
- 고정된 설정값
- 함수에서 여러 값 반환할 때

---

````python
##################################################
# 튜플 사용하기
##################################################

# 튜플 생성
point = (10, 20)  # 2D 좌표
rgb = (255, 0, 0)  # 빨간색
person = ("김철수", 25, "서울")

# 인덱싱과 슬라이싱 (리스트처럼)
print(point[0])    # 10
print(rgb[1:])     # (0, 0)

# 언패킹 (값 꺼내기)
x, y = point
print(x, y)  # 10 20

name, age, city = person
print(f"{name}은 {age}살이고 {city}에 살아요")

# 튜플은 수정 불가!
# point[0] = 30  # ❌ 에러!
# point.append(30)  # ❌ 에러!

# 하나짜리 튜플 (쉼표 필수!)
single = (5,)   # 튜플
not_tuple = (5)  # 그냥 숫자
````

## 💡 핵심 포인트

### 1️⃣ 튜플은 수정 불가능 (Immutable)
```python
point = (10, 20)
point[0] = 30  # ❌ 에러 발생!
```

### 2️⃣ 언패킹 (Unpacking)
튜플의 값을 여러 변수에 한 번에 할당!
```python
x, y = (10, 20)
# x = 10, y = 20
```

### 3️⃣ 하나짜리 튜플은 쉼표 필수!
```python
single = (5,)   # ✅ 튜플
not_tuple = (5) # ❌ 그냥 숫자 5
```

### 4️⃣ 인덱싱/슬라이싱은 리스트처럼!
```python
point[0]   # 첫 번째 값
rgb[1:]    # 1번부터 끝까지
```

---

## 🚀 튜플을 사용하는 이유

### ✅ 장점
1. **안전성**: 실수로 값을 바꿀 수 없어요
2. **속도**: 리스트보다 빠름
3. **메모리**: 적게 사용

### 📌 실전 예시

**좌표 데이터**
```python
point = (10, 20)  # 변하지 않는 위치
```

**색상 코드**
```python
red = (255, 0, 0)    # 빨강
green = (0, 255, 0)  # 초록
blue = (0, 0, 255)   # 파랑
```

**사용자 정보 (고정)**
```python
user = ("user123", "김철수", 25)
user_id, name, age = user  # 언패킹
```
